```python
import pandas as pd
import matplotlib.pyplot as plt

# Epoch values in nanoseconds for 2nd, 3rd and 4th-Sept-2025
epoch_ns = [1756794461007837000, 1756894461007837000, 1756994461007837000]

# Converting epoch to datetime
datetimes = pd.to_datetime(epoch_ns, unit="ns")

# Converting datetime to date
dates = datetimes.date

# Creating a DataFrame
df = pd.DataFrame({
    "epoch_ns": epoch_ns,
    "datetime": datetimes,
    "date": dates,
    "value": [10, 20, 15]   # values for 3 dates
})

print(df)

# Visualization: time series plot
plt.figure(figsize=(8,4))
plt.plot(df["datetime"], df["value"], marker="o", linestyle="-")
plt.title("Epoch â†’ Datetime Conversion with Values")
plt.xlabel("Datetime")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
